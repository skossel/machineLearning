# Decision Tree Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

## Importing the dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Decision Tree Regression model on the whole dataset

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

## Predicting a new result

```python
regressor.predict([[6.5]])
```

## Visualising the Decision Tree Regression results (higher resolution)

```python
# Stufenartige Kurve: weil wir den Decision Tree verwenden. Dieser trennt und Yes oder No und deshalb solch eine ruckartige Kurve.

# erzeugt mehr Datenpunkte, um eine glatte Linie zu plotten
X_grid = np.arange(X.min(), X.max(), 0.1)
# macht aus 1D Array von Datenpunkten ein 2D Array [[Datenpunkte]]
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
